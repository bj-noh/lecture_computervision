# OpenCV Practice Lab

This notebook reorganizes the `#2 OpenCV practice` lecture into a guided lab.

Topics:
- image loading and inspection
- grayscale conversion and resizing
- drawing primitives and annotations
- a small applied summary image

```python
from pathlib import Path

import cv2
import matplotlib.pyplot as plt
import numpy as np

BASE_DIR = Path.cwd()
if not (BASE_DIR / "data").exists():
    BASE_DIR = Path("lab-session/01-opencv-practice")

DATA_DIR = BASE_DIR / "data"
OUTPUT_DIR = BASE_DIR / "outputs"
OUTPUT_DIR.mkdir(exist_ok=True)

IMAGE_PATH = DATA_DIR / "cat.jpg"
print("Base dir:", BASE_DIR.resolve())
print("Image path:", IMAGE_PATH.resolve())
```

```python
image_bgr = cv2.imread(str(IMAGE_PATH))
if image_bgr is None:
    raise FileNotFoundError(f"Could not open image: {IMAGE_PATH}")

image_rgb = cv2.cvtColor(image_bgr, cv2.COLOR_BGR2RGB)
print("Image shape:", image_bgr.shape)

plt.figure(figsize=(6, 4))
plt.imshow(image_rgb)
plt.title("Original image")
plt.axis("off");
```

## Transform: grayscale and resize

The lecture converts a color image into grayscale and then changes the spatial size.

```python
gray = cv2.cvtColor(image_bgr, cv2.COLOR_BGR2GRAY)
gray_small = cv2.resize(gray, dsize=(0, 0), fx=0.5, fy=0.5)

cv2.imwrite(str(OUTPUT_DIR / "cat_gray.jpg"), gray)
cv2.imwrite(str(OUTPUT_DIR / "cat_gray_small.jpg"), gray_small)

fig, axes = plt.subplots(1, 2, figsize=(10, 4))
axes[0].imshow(gray, cmap="gray")
axes[0].set_title("Gray")
axes[0].axis("off")
axes[1].imshow(gray_small, cmap="gray")
axes[1].set_title("Gray small")
axes[1].axis("off")
plt.tight_layout()
print("Gray shape:", gray.shape)
print("Resized shape:", gray_small.shape)
```

## Drawing primitives

OpenCV provides line, rectangle, circle, and text drawing utilities that are frequently used for visualization.

```python
annotated = image_bgr.copy()
cv2.line(annotated, (20, 20), (200, 160), (255, 0, 0), 4)
cv2.rectangle(annotated, (220, 40), (380, 180), (0, 255, 0), 3)
cv2.circle(annotated, (130, 270), 55, (0, 0, 255), -1)
cv2.putText(
    annotated,
    "OpenCV Lab",
    (210, 280),
    cv2.FONT_HERSHEY_SIMPLEX,
    0.9,
    (255, 255, 255),
    2,
    cv2.LINE_AA,
)

annotated_rgb = cv2.cvtColor(annotated, cv2.COLOR_BGR2RGB)
cv2.imwrite(str(OUTPUT_DIR / "cat_annotated.jpg"), annotated)

plt.figure(figsize=(6, 4))
plt.imshow(annotated_rgb)
plt.title("Annotated image")
plt.axis("off");
```

## Applied example: summary board

This cell creates a compact result image that combines original, grayscale, edge, and annotated views.

```python
edges = cv2.Canny(gray, 80, 160)

tile_size = (320, 240)
original_tile = cv2.resize(image_bgr, tile_size)
gray_tile = cv2.cvtColor(cv2.resize(gray, tile_size), cv2.COLOR_GRAY2BGR)
edge_tile = cv2.cvtColor(cv2.resize(edges, tile_size), cv2.COLOR_GRAY2BGR)
annotated_tile = cv2.resize(annotated, tile_size)

def label(tile, text):
    out = tile.copy()
    cv2.rectangle(out, (0, 0), (160, 34), (30, 30, 30), -1)
    cv2.putText(out, text, (10, 24), cv2.FONT_HERSHEY_SIMPLEX, 0.7, (255, 255, 255), 2, cv2.LINE_AA)
    return out

top = np.hstack([label(original_tile, "Original"), label(gray_tile, "Gray")])
bottom = np.hstack([label(edge_tile, "Edges"), label(annotated_tile, "Annotated")])
summary = np.vstack([top, bottom])

summary_path = OUTPUT_DIR / "cat_summary_board.jpg"
cv2.imwrite(str(summary_path), summary)

plt.figure(figsize=(10, 7))
plt.imshow(cv2.cvtColor(summary, cv2.COLOR_BGR2RGB))
plt.title("Summary board")
plt.axis("off")
print("Saved summary board to:", summary_path.resolve())
```
